```python
# ============================
# Dependencies
# ============================
import numpy as np                       # numerical ops, arrays, linear algebra
import pandas as pd                      # data loading/manipulation
from scipy import sparse                 # memory-efficient sparse matrices
from sklearn.feature_extraction.text import CountVectorizer  # bag-of-ngrams
from sklearn.manifold import MDS         # multidimensional scaling for 2D layout
from IPython.display import display      # nicer table display in notebooks
import matplotlib.pyplot as plt          # plotting

# ============================
# Config (edit these easily)
# ============================
# CSV source (read-only from GitHub; works anywhere, incl. Colab)
CSV_PATH = "https://raw.githubusercontent.com/AHMerrill/Unstructured-Data-1/main/car_comments_zm.csv"
COMMENT_COL            = "comment"  # text column in the CSV
TOP_BRANDS_BY_COUNT    = 9          # how many top brands by raw mention count to always include
TOP_PAIRS_BY_COMENTION = 10         # how many brand pairs to show by co-mention count
TOP_PAIRS_BY_LIFT      = 10         # how many brand pairs to show by lift (association strength)
MIN_COMENTION_SUPPORT  = 300        # ignore weak pairs (< this many co-mentions)
K_SMOOTH               = 1.0        # add-k smoothing for lift to avoid divide-by-zero
RANDOM_STATE           = 42         # seed for reproducible MDS
EPS                    = 1e-12      # tiny epsilon for safe divisions / inverses

# ============================
# Preconditions
# ============================
# You must define `brand_aliases` beforehand.
# Expected shape: { "tesla": ["tesla","tsla","model 3","m3",...], "bmw":[...], ... }
try:
    brand_aliases
except NameError:
    raise RuntimeError("Define `brand_aliases` (dict) before running this cell.")

# ============================
# 1) Load comments (already clean per your pipeline)
# ============================
car_comments_df = pd.read_csv(CSV_PATH)              # load comments CSV
comments = car_comments_df[COMMENT_COL]              # select text column
comments = comments.fillna("").astype(str)           # ensure string, replace NaN
N = int(len(comments))                               # total number of comments (docs)
if N == 0:
    raise ValueError("No comments found in CSV.")

# ============================
# 2) alias → brand map (lowercased)
#    Build a lookup from any alias/token to its canonical brand name.
# ============================
alias_to_brand = {}
for brand, alias_list in brand_aliases.items():
    b = str(brand).strip().lower()                   # canonical brand (lowercase)
    for alias in alias_list:
        a = str(alias).strip().lower()               # alias token (lowercase)
        if a and a not in alias_to_brand:            # first alias wins if duplicated
            alias_to_brand[a] = b

# List of canonical brands and a stable index for each
brands = sorted(set(alias_to_brand.values()))
brand_to_idx = {b: i for i, b in enumerate(brands)}

# ============================
# 3) Vectorize on aliases (1–4 grams, fixed vocab)
#    We count occurrences of *aliases* with a fixed vocabulary so
#    the vectorizer doesn't learn anything outside our alias list.
# ============================
alias_vocab = sorted(alias_to_brand.keys())          # all alias tokens we care about
vocab = {a: i for i, a in enumerate(alias_vocab)}    # alias -> column index in X_alias

# CountVectorizer config:
# - lowercase=True: normalize text
# - token_pattern uses word tokens (keeps letters/numbers/_)
# - ngram_range=(1,4): capture multiword aliases like "land rover" or "mercedes benz amg"
vectorizer = CountVectorizer(
    vocabulary=vocab,            # fixed vocab = our aliases only
    lowercase=True,
    token_pattern=r"(?u)\b\w+\b",
    ngram_range=(1, 4),
)

# Produce sparse matrix of alias counts per document: shape (N_docs x N_aliases)
X_alias = vectorizer.transform(comments.values)

# ============================
# 4) Collapse aliases → brands, then binarize per comment
#    Map each alias column to its brand and mark presence/absence.
# ============================
# Build a (N_aliases x N_brands) sparse matrix A where A[alias, brand]=1 if alias maps to brand.
rows, cols = [], []
for alias, b in alias_to_brand.items():
    rows.append(vocab[alias])          # alias column index in X_alias
    cols.append(brand_to_idx[b])       # brand column index in X_brand

A = sparse.csr_matrix(
    (np.ones(len(rows), dtype=np.int8), (rows, cols)),
    shape=(len(alias_vocab), len(brands))
)

# Multiply: (N_docs x N_aliases) @ (N_aliases x N_brands) -> (N_docs x N_brands) alias counts per brand
X_brand_counts = X_alias @ A

# Binarize: did the comment mention the brand at least once? (1/0)
X = (X_brand_counts > 0).astype(np.int8).tocsr()

# ============================
# 5) Co-mentions & marginals
#    Compute brand-by-brand co-occurrence matrix C and brand marginals.
# ============================
X32 = X.astype(np.int32)               # prevent int8 overflow in multiplication
C = X32.T @ X32                        # (n_brands x n_brands): C[i,j] = #docs mentioning both brand i and j
C_dense = C.toarray().astype(np.int64) # dense for simple math/printing
counts = np.asarray(C.diagonal()).astype(np.int64)  # counts[i] = #docs mentioning brand i

# ============================
# 6) Lift (market-basket) with optional add-k smoothing
#    lift(A,B) = N * ( #(A,B) + k ) / ( ( #(A) + k ) * ( #(B) + k ) )
#    Intuition: >1 means A and B co-occur more than expected if independent.
# ============================
k = float(K_SMOOTH)                    # smoothing for stability on rare counts
Aplus = (counts[:, None] + k).astype(np.float64)     # #(A)+k as column
Bplus = (counts[None, :] + k).astype(np.float64)     # #(B)+k as row
ABplus = (C_dense + k).astype(np.float64)            # #(A,B)+k

with np.errstate(divide="ignore", invalid="ignore"):  # hush divide-by-zero warnings
    lift = (N * ABplus) / (Aplus * Bplus + EPS)      # lift matrix
    lift[~np.isfinite(lift)] = 0.0                   # set NaN/inf to 0

# ============================
# 7) Per-brand counts + top pair tables
# ============================
# Table: how many docs mention each brand
counts_df = pd.DataFrame({"brand": brands, "count": counts}).sort_values("count", ascending=False)
print("Top brands by mentions:\n")
display(counts_df.head(TOP_BRANDS_BY_COUNT))

# Build one table of all brand pairs: (brand_a, brand_b, co_mentions, lift)
pairs = []
for i in range(len(brands)):
    for j in range(i + 1, len(brands)):              # upper triangle only; (A,B) == (B,A)
        co = int(C_dense[i, j])                      # co-mentions
        if co > 0:
            pairs.append((brands[i], brands[j], co, float(lift[i, j])))

pairs_df = pd.DataFrame(pairs, columns=["brand_a", "brand_b", "co_mentions", "lift"])

# Filter out weak pairs, then show top-K by co-mentions (tie-break by lift)
pairs_supported = pairs_df[pairs_df["co_mentions"] >= MIN_COMENTION_SUPPORT].copy()
top_by_co = (
    pairs_supported.sort_values(["co_mentions", "lift"], ascending=[False, False])
    .head(TOP_PAIRS_BY_COMENTION)
    .reset_index(drop=True)
)
print(f"\nTop {TOP_PAIRS_BY_COMENTION} pairs by co-mentions (min support = {MIN_COMENTION_SUPPORT}):")
display(top_by_co)

# Top-K by lift among supported pairs (strongest associations)
top_by_lift = (
    pairs_supported.sort_values("lift", ascending=False)
    .head(TOP_PAIRS_BY_LIFT)
    .reset_index(drop=True)
)
print(f"\nTop {TOP_PAIRS_BY_LIFT} pairs by lift (min support = {MIN_COMENTION_SUPPORT}):")
display(top_by_lift)

# ============================
# 8) MDS plot: pick a brand subset, build a graph distance, embed to 2D
#    We include brands from:
#      - Top-N by mentions
#      - Top-K by co-mentions
#      - Top-K by lift
#    Edges come from BOTH top-K lists; distance is inverse-lift (higher lift = closer).
# ============================

# --- Collect the brand set to plot
top_mentions_brands = counts_df.head(TOP_BRANDS_BY_COUNT)["brand"].tolist()
brands_from_co     = sorted(set(top_by_co["brand_a"]).union(set(top_by_co["brand_b"])))
brands_from_lift   = sorted(set(top_by_lift["brand_a"]).union(set(top_by_lift["brand_b"])))
brands_in_mds = sorted(set(top_mentions_brands) | set(brands_from_co) | set(brands_from_lift))

# Map brand names to indices in (a) our subset and (b) the full matrices
name_to_idx_local  = {b: i for i, b in enumerate(brands_in_mds)}
name_to_idx_master = {b: brands.index(b) for b in brands_in_mds}

# Slice the lift matrix down to only the selected brands
lift_sub = lift[
    np.ix_(
        [name_to_idx_master[b] for b in brands_in_mds],
        [name_to_idx_master[b] for b in brands_in_mds],
    )
]

# Initialize a distance matrix with "infinite" distances (no edges yet)
K_nodes = len(brands_in_mds)
D = np.full((K_nodes, K_nodes), np.inf, dtype=float)
np.fill_diagonal(D, 0.0)  # distance from a node to itself is 0

# Build the edge list from top-K-by-lift and top-K-by-co-mentions (undirected)
edges = pd.concat(
    [top_by_lift[["brand_a", "brand_b"]], top_by_co[["brand_a", "brand_b"]]],
    ignore_index=True
).drop_duplicates()

# For every edge (a,b), set distance = 1/max(lift, EPS). Higher lift -> shorter distance.
for _, row in edges.iterrows():
    a, b = row["brand_a"], row["brand_b"]
    if a in name_to_idx_local and b in name_to_idx_local:
        i = name_to_idx_local[a]
        j = name_to_idx_local[b]
        lij = float(lift_sub[i, j])
        d = 1.0 / max(lij, EPS)
        # If multiple edges propose a distance, keep the shortest
        D[i, j] = min(D[i, j], d)
        D[j, i] = min(D[j, i], d)

# All-pairs shortest paths (Floyd–Warshall) to fill in indirect distances via graph paths
for k_idx in range(K_nodes):
    D = np.minimum(D, D[:, [k_idx]] + D[[k_idx], :])

# If any nodes are still disconnected (inf), set them to a large but finite distance
if np.isinf(D).any():
    finite_vals = D[np.isfinite(D) & (D > 0)]
    big = (np.median(finite_vals) if finite_vals.size else 1.0) * 10.0
    D[np.isinf(D)] = big

# Normalize distances to stabilize MDS scale (divide by median off-diagonal distance)
off = ~np.eye(K_nodes, dtype=bool)
median_d = np.median(D[off]) if np.any(off) else 1.0
if median_d > 0:
    D = D / median_d

# Run classical MDS with our precomputed graph distances to get 2D coordinates
mds = MDS(
    n_components=2,
    dissimilarity="precomputed",
    random_state=RANDOM_STATE,
    n_init=6,
    max_iter=1000,
)
coords = mds.fit_transform(D)  # shape (K_nodes, 2)

# ============================
# Plot (scatter + offset labels to reduce overplotting)
# ============================
plt.figure(figsize=(9, 8))
plt.scatter(coords[:, 0], coords[:, 1], s=30)

# Center & scale info for placing label offsets radially away from the center
cx, cy = coords[:, 0].mean(), coords[:, 1].mean()
rx = max(np.ptp(coords[:, 0]), 1e-9)  # NumPy 2.x: use np.ptp instead of arr.ptp()
ry = max(np.ptp(coords[:, 1]), 1e-9)

for i, name in enumerate(brands_in_mds):
    dx, dy = coords[i, 0] - cx, coords[i, 1] - cy
    norm = np.hypot(dx, dy) + 1e-9
    # Offset labels slightly away from points along the radial direction
    ox = (dx / norm) * 0.025 * rx
    oy = (dy / norm) * 0.025 * ry
    plt.annotate(
        name,
        (coords[i, 0], coords[i, 1]),
        xytext=(coords[i, 0] + ox, coords[i, 1] + oy),
        textcoords="data",
        arrowprops=dict(arrowstyle="-", lw=0.5),
        fontsize=9, ha="center", va="center",
        bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="none", alpha=0.7),
    )

plt.title(
    f"MDS of brands from Top-{TOP_BRANDS_BY_COUNT} mentions, "
    f"Top-{TOP_PAIRS_BY_COMENTION} co-mentions & Top-{TOP_PAIRS_BY_LIFT} lift"
)
plt.xticks([]); plt.yticks([])
plt.tight_layout()
plt.show()
```

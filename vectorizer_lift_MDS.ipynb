```python
# ============================
# Dependencies
# ============================
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import MDS
from IPython.display import display
import matplotlib.pyplot as plt

# ============================
# Config (edit these easily)
# ============================
CSV_PATH              = "car_comments.csv"
COMMENT_COL           = "comment"
TOP_BRANDS_BY_COUNT   = 9       # <-- top-N brands by mentions
TOP_PAIRS_BY_COMENTION= 10      # <-- top-K pairs by co-mentions
TOP_PAIRS_BY_LIFT     = 10      # <-- top-K pairs by lift (used for MDS)
MIN_COMENTION_SUPPORT = 300     # <-- ignore pairs with fewer co-mentions than this
K_SMOOTH              = 1.0     # <-- add-k smoothing for lift; set 0 to disable
RANDOM_STATE          = 42      # useful for MDS recreation
EPS                   = 1e-12   # tiny epsilon for safe division

# ============================
# Preconditions
# ============================
try:
    brand_aliases
except NameError:
    raise RuntimeError("Define `brand_aliases` (dict) before running this cell.")

# ============================
# 1) Load comments (already clean per your pipeline)
# ============================
car_comments_df = pd.read_csv(CSV_PATH)
comments = car_comments_df[COMMENT_COL].fillna("").astype(str)
N = int(len(comments))
if N == 0:
    raise ValueError("No comments found in CSV.")

# ============================
# 2) alias → brand map (lowercased)
# ============================
alias_to_brand = {}
for brand, alias_list in brand_aliases.items():
    b = str(brand).strip().lower()
    for alias in alias_list:
        a = str(alias).strip().lower()
        if a and a not in alias_to_brand:
            alias_to_brand[a] = b

brands = sorted(set(alias_to_brand.values()))
brand_to_idx = {b: i for i, b in enumerate(brands)}

# ============================
# 3) Vectorize on aliases (1–4 grams, fixed vocab)
# ============================
alias_vocab = sorted(alias_to_brand.keys())
vocab = {a: i for i, a in enumerate(alias_vocab)}
vectorizer = CountVectorizer(
    vocabulary=vocab,
    lowercase=True,
    token_pattern=r"(?u)\b\w+\b", # uses words as tokens
    ngram_range=(1, 4),
)
X_alias = vectorizer.transform(comments.values)   # (N x n_aliases) CSR

# ============================
# 4) Collapse aliases → brands, binarize per comment
# ============================
rows, cols = [], []
for alias, b in alias_to_brand.items():
    rows.append(vocab[alias])
    cols.append(brand_to_idx[b])

A = sparse.csr_matrix(
    (np.ones(len(rows), dtype=np.int8), (rows, cols)),
    shape=(len(alias_vocab), len(brands))
)
X_brand_counts = X_alias @ A                                 # (N x n_brands) counts
X = (X_brand_counts > 0).astype(np.int8).tocsr()             # binary presence

# ============================
# 5) Co-mentions & marginals
# ============================
X32 = X.astype(np.int32)                      # avoid int8 overflow
C = X32.T @ X32                               # (n_brands x n_brands) CSR
C_dense = C.toarray().astype(np.int64)        # #(A,B)
counts = np.asarray(C.diagonal()).astype(np.int64)  # #(A)

# ============================
# 6) Lift (market-basket) with optional add-k smoothing
#     lift = N * ( #(A,B) + k ) / ( ( #(A) + k ) ( #(B) + k ) )
# ============================
k = float(K_SMOOTH)
Aplus = (counts[:, None] + k).astype(np.float64)
Bplus = (counts[None, :] + k).astype(np.float64)
ABplus = (C_dense + k).astype(np.float64)

with np.errstate(divide="ignore", invalid="ignore"):
    lift = (N * ABplus) / (Aplus * Bplus + EPS)
    lift[~np.isfinite(lift)] = 0.0

# ============================
# 7) Per-brand counts + tables
# ============================
counts_df = pd.DataFrame({"brand": brands, "count": counts}).sort_values("count", ascending=False)
print("Top brands by mentions:\n")
display(counts_df.head(TOP_BRANDS_BY_COUNT))

# Build full pairs table once
pairs = []
for i in range(len(brands)):
    for j in range(i + 1, len(brands)):
        co = int(C_dense[i, j])
        if co > 0:
            pairs.append((brands[i], brands[j], co, float(lift[i, j])))

pairs_df = pd.DataFrame(pairs, columns=["brand_a", "brand_b", "co_mentions", "lift"])

# Top-K by co-mentions (with support filter)
pairs_supported = pairs_df[pairs_df["co_mentions"] >= MIN_COMENTION_SUPPORT].copy()
top_by_co = pairs_supported.sort_values(["co_mentions", "lift"], ascending=[False, False]) \
                           .head(TOP_PAIRS_BY_COMENTION) \
                           .reset_index(drop=True)

print(f"\nTop {TOP_PAIRS_BY_COMENTION} pairs by co-mentions (min support = {MIN_COMENTION_SUPPORT}):")
display(top_by_co)

# Top-K by lift (within support-filtered pairs)
top_by_lift = pairs_supported.sort_values("lift", ascending=False) \
                             .head(TOP_PAIRS_BY_LIFT) \
                             .reset_index(drop=True)

print(f"\nTop {TOP_PAIRS_BY_LIFT} pairs by lift (min support = {MIN_COMENTION_SUPPORT}):")
display(top_by_lift)

# ============================
# 8) MDS including brands from:
#    - Top-N by mentions
#    - Top-K by co-mentions (after support)
#    - Top-K by lift (after support)
#    Edges come from BOTH Top-K-by-lift and Top-K-by-co-mentions.
# ============================

# --- collect brand sets we must include in the MDS ---
top_mentions_brands = counts_df.head(TOP_BRANDS_BY_COUNT)["brand"].tolist()
brands_from_co     = sorted(set(top_by_co["brand_a"]).union(set(top_by_co["brand_b"])))
brands_from_lift   = sorted(set(top_by_lift["brand_a"]).union(set(top_by_lift["brand_b"])))

# union of all three sets
brands_in_mds = sorted(set(top_mentions_brands) | set(brands_from_co) | set(brands_from_lift))

# local <-> master index maps
name_to_idx_local  = {b: i for i, b in enumerate(brands_in_mds)}
name_to_idx_master = {b: brands.index(b) for b in brands_in_mds}

# extract lift submatrix for these brands
lift_sub = lift[np.ix_([name_to_idx_master[b] for b in brands_in_mds],
                       [name_to_idx_master[b] for b in brands_in_mds])]

# initialize distances (graph, start with no edges)
K_nodes = len(brands_in_mds)
D = np.full((K_nodes, K_nodes), np.inf, dtype=float)
np.fill_diagonal(D, 0.0)

# --- build edge list from BOTH top-k tables ---
edges = pd.concat([
    top_by_lift[["brand_a", "brand_b"]],
    top_by_co[["brand_a", "brand_b"]],
], ignore_index=True).drop_duplicates()

# add undirected edges with inverse-lift distance
for _, row in edges.iterrows():
    a, b = row["brand_a"], row["brand_b"]
    if a in name_to_idx_local and b in name_to_idx_local:
        i = name_to_idx_local[a]
        j = name_to_idx_local[b]
        lij = float(lift_sub[i, j])
        d   = 1.0 / max(lij, EPS)     # higher lift => shorter distance
        D[i, j] = min(D[i, j], d)
        D[j, i] = min(D[j, i], d)

# Floyd–Warshall: all-pairs shortest paths on this small graph
for k_idx in range(K_nodes):
    D = np.minimum(D, D[:, [k_idx]] + D[[k_idx], :])

# if disconnected components remain, fill with a large finite distance
if np.isinf(D).any():
    finite_vals = D[np.isfinite(D) & (D > 0)]
    big = (np.median(finite_vals) if finite_vals.size else 1.0) * 10.0
    D[np.isinf(D)] = big

# normalize distances to stabilize MDS
off = ~np.eye(K_nodes, dtype=bool)
median_d = np.median(D[off]) if np.any(off) else 1.0
if median_d > 0:
    D = D / median_d

# run MDS
mds = MDS(n_components=2, dissimilarity="precomputed",
          random_state=RANDOM_STATE, n_init=6, max_iter=1000)
coords = mds.fit_transform(D)

# plot with offset labels, no tick labels
plt.figure(figsize=(9, 8))
plt.scatter(coords[:, 0], coords[:, 1], s=30)

cx, cy = coords[:, 0].mean(), coords[:, 1].mean()
rx = max(coords[:, 0].ptp(), 1e-9)
ry = max(coords[:, 1].ptp(), 1e-9)

for i, name in enumerate(brands_in_mds):
    dx, dy = coords[i, 0] - cx, coords[i, 1] - cy
    norm = np.hypot(dx, dy) + 1e-9
    ox = (dx / norm) * 0.025 * rx
    oy = (dy / norm) * 0.025 * ry
    plt.annotate(
        name,
        (coords[i, 0], coords[i, 1]),
        xytext=(coords[i, 0] + ox, coords[i, 1] + oy),
        textcoords="data",
        arrowprops=dict(arrowstyle="-", lw=0.5),
        fontsize=9, ha="center", va="center",
        bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="none", alpha=0.7),
    )

plt.title(
    f"MDS of brands from Top-{TOP_BRANDS_BY_COUNT} mentions, "
    f"Top-{TOP_PAIRS_BY_COMENTION} co-mentions & Top-{TOP_PAIRS_BY_LIFT} lift"
)
plt.xticks([]); plt.yticks([])
plt.tight_layout()
plt.show()
```
